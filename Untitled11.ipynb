```python
X = diabetes['BMI'].values
y = diabetes['DiabetesPedigreeFunction'].values
```

```python
plt.scatter(X,y , alpha=0.7, edgecolors = 'w')
plt.title('Relationship between BMI and Diabetes ')
plt.xlabel('BMI ( Body Mass Index)')
plt.ylabel('Diabetes Progression')
plt.show()
```

```python
import scipy.stats as stats
```

```python
# Compute the linear regression parameters
slope, intercept, r_value, p_value, std_err = stats.linregress(X,y)

# Print the results
print("Slope:",slope)
print("Intercept:",intercept)
print("R-squared:", r_value**2)
```

```python
plt.scatter(X,y, alpha=0.7, edgecolors='w')
plt.plot(X, slope * X + intercept, color='red', linewidth=2 )


plt.title('Simple Linear Regression  ')
plt.xlabel('BMI ( Body Mass Index)')
plt.ylabel('Diabetes Progression')
plt.show()
```

```python
X = diabetes.drop('DiabetesPedigreeFunction', axis=1).values
y = diabetes['DiabetesPedigreeFunction'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
print("Inteception:",model.intercept_)
print("Coefficients:",model.coef_)
```

```python
y_pred = model.predict(X)
y_pred
```

```python
model.score(X,y)
```
